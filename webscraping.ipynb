```python
#Web sreaping es un metodo que me permite tomar información que está disponible en la web y
#Asi funciona gran parte del modelo de gpt
#Esto no es una practica ilegal
import pandas as pd
```

```python
#Hay muchas formas de hacer web scraping sin embargo vamos a hacerlo simple
datos_futbol=pd.read.csv("https://www.football-data.co.uk/mmz4281/2425/E0.csv")
```

```python
#Ahora le voy a cambiar los nombres a las columnas oara poder verlo de una mejor manera
datos_futbol.rename(columns={"fecha":"Fecha",
                             "HomeTeam":"Local",
                             "AwayTeam":"Visitante",
                             "FTNG":"Goles Locales",
                             "FTAG":"Goles Visitantes"})
```

```python
#La siguiente es la estructura que podemos ver en el link
"https://www.football-data.co.uk/mmz4281/"+"2425"+"/"+"E0"+".csv"
```

```python
#Ya se qye la raiz es la misma entonces la guardo en una variable que se llame ruta
ruta="https://www.football-data.co.uk/mmz4281/"
```

```python
#Ahora lo que voy a hacer es crearme una lista con los compeonatos o terminaciones
campeonatos=["E0","E1","E2","E3","EC"]
enlaces=[]
#Para que esto funcione, debo recorrer la lista con un loop, a esta práctica se le conoce como
#Link dinámico, una vez recorra la lista, debo reemplazar el valor para qie el enlace funcione

for campeonato in campeonatos:
  df_links=pd.read_csv(ruta + "2425" + "/" + campeonato + ".csv" )
  enlaces.append(df_links)
```
