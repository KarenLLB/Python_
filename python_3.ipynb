```python
#siempre ño primero que debo hacer es importar pandas y numpy
import pandas as pd
import numpy as np
```

```python
#Ahora subimos el archivo desde google drive, ya no tenemos que loguearnos de nuevo
df_estudiantes=pd.read_csv("/content/drive/MyDrive/Python_Curso/StudentsPerformance (1).csv")
df_estudiantes
```

```python
#Sigamos con la creacion de columnas usando el metodo assign o el metodo insert
#Si quiero agregar varias columnas uso assign y tambirn si quiero sobreescribir una columna
```

```python
#Freamos notas de disica y filosofia aleatorias y las guardamos en variables
filosofia_score=np.random.randint(1,100,size=1000)
fisica_score=np.random.randint(1,100,size=1000)
```

```python
#Ahora le pondre un indice a cada fila para que sepa donde va en el dataframe
filososfia_nota=pd.Series(filosofia_score, index=np.arange(0,1000))
fisica_nota=pd.Series(fisica_score, index=np.arange(0,1000))
```

```python
#Ahora lo que voy a hacer es poner las nuevas columnas a mi dataframe, usando assign
df_estudiantes.assign(filosofia_score=filososfia_nota, fisica_score=fisica_nota)
```

```python
df_estudiantes=df_estudiantes.assign(filosofia_score=filososfia_nota, fisica_score=fisica_nota)
df_estudiantes
```

```python
#Ahora vamos a usar el metodo insert, el metodo insert es diferente porque me permite especificar donde yo voy a colocar esa columan
#Al usar el metodfo insert debemos tener cuidado porque no me devuelve copias sino el cambio queda hecho de una vez
df_estudiantes.insert(1, "ED_SCORE", fisica_nota)
```
