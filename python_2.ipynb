Para poder usar dataframe  vamos a usar dos bibliotecas, una de ellas es numpy y la otra es pandas, las dos se complementan y, hacen cosas diferentes.Para crear dataframe existen varias posibilidades, sin embargo, la más usada es la lectura directa desde un archivo.

```python
import pandas as pd #Importo pansas y le pongo el sobrenombre pd
import numpy as np #Importo numpay y le pongo el sobrenombre np
```

```python
# La primera forma de crear un dataframe es usando el metodo array, y luego guardarlo en una variable.
dataframe_1= np.array([[2,4],[6,8],[10,12],[14,16]])
```

```python
#Para crear el dataframe del arreglo anterior, debemos usar pandas
pd.DataFrame(dataframe_1)
```

```python
dataframe_2= np.array([[3,6,9],[12,15,18],[21,24,27],[30,33,36]])
```

```python
pd.DataFrame(dataframe_2)
```

```python
#Ahora vamos a cambiarle el nombre a las columnas, esto se hace de la siguiente manera
df1=pd.DataFrame(dataframe_1, index=["fila1","fila2","fila3","fila4"],
                 columns=["columna1","columna2"])
```

```python
df1
```

```python
#Ahora creamos un dataframe usando listas
dataframe_lista=[("Karen","Lagos",24),("Carlos", "Lagos",61),("Liliana","Bautista",54),("Hortencia","Lagos",79),("Angie","Lagos",30),("Camilo","Bautista",30)]
```

```python
df4=pd.DataFrame(dataframe_lista, index=["1","2","3","4","5","6"],
                     columns=["Nombre","Apellido","Edad"])
```

```python
df4
```

```python
#Primero creamos dos listas puedo crear cuantas listas quiera, entiendo que las listas son las columnas del dataframe
ciudadColombia=["Bogotá","Medellin","Cali","Barranquilla"]
ciudadesInfo=[11000000,2500000,2300000,1300000]
```

```python
#Ahora creamos los diccionarios de las listas
ciudades={"ciudades":ciudadColombia,"población":ciudadesInfo}
```

```python
#Le paso el metodo de pandas
df5=pd.DataFrame(ciudades)
```

```python
#llamo el Dataframe
df5
```

```python
#Hacer dataframe de 4 filas por 3 columnas de 4 paises
paisesMundo=["Colombia","Estados Unidos","Italia","Turquia"]
poblacionMundo=[51000000,335893238,58990000,86224025]
expectativavidaMundo=[73,81,84,77]
```

```python
paises={"paises":paisesMundo,"población":poblacionMundo,"expectativa":expectativavidaMundo}
```

```python
df6=pd.DataFrame(paises)
```
